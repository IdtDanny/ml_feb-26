```python
import matplotlib.pyplot as plt
import pandas as pd

data = {'TransactionAmount': [50,75,100,45,60,55,1200,80,90,3000,655]}
df = pd.DataFrame(data)

print(f'{df}')

plt.boxplot(df['TransactionAmount'])
plt.title('Transaction Amount BoxPlot')
plt.ylabel('Amount')
plt.show()
```
