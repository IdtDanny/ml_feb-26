```python
# Treating the missing data
import pandas as pd
import numpy as np

data = {
    'Age':[25,30,np.nan,22,28,np.nan],
    'Salary':[50000,60000,55000,np.nan,52000,58000],
    'Department':['HR','Finane',None,'IT','Finance','HR'],
    'Gender': ['Male','Female','Female',None,'Male','Female']
}
df=pd.DataFrame(data)

df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(), inplace = True)
df['Salary'].fillna(df['Salary'].median(), inplace = True)

dep_mode = df['Department'].mode()[0]
ge_mode = df['Gender'].mode()[0]

df['Department'].fillna(dep_mode, inplace = True)
df['Gender'].fillna(ge_mode, inplace = True)

df.isnull().sum()
```
