```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random

data = {
    'Age': [25, 30, np.nan, 22, 28, np.nan],
    'Salary': [50000, 60000, 55000, np.nan, 52000, 58000],
    'Department': ['HR', 'Finance', None, 'IT', 'Finance', 'HR'],
    'Gender': ['Male', 'Female', 'Female', None, 'Male', 'Female']
}

df = pd.DataFrame(data)

print(df)
print(f'\nChecking Missing Values: \n{df.isnull().sum()}')
```

```python
print(f'The first 6 rows are: \n {df.head(6)}')
```

```python
df['Salary'].fillna(df['Salary'].mean(), inplace = True)
print(f'Handling Missing Salary Data with Mean: \n{df}')
print(f'\nMissing Data: \n{df.isnull().sum()}')

print(f'\nAverage Salary: {df['Salary'].mean()}')
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
dep_mode = df['Department'].mode()[0]
ge_mode = df['Gender'].mode()[0]
df['Department'].fillna(dep_mode, inplace=True)
df['Gender'].fillna(ge_mode, inplace=True)
print(f'Summary on Missing Data: \n{df.isnull().sum()}')

# Adding Random satisfactory Score
rand = []
for i in range(6):
    rand.append(random.randint(1,10))
df['Score'] = rand
print(f'\nAfter Adding Random Score: \n{df}')

x = df['Age']
y = df['Score']
plt.scatter(x,y, color='green')
plt.title('Relation between Age and Score', fontsize = 16)
plt.xlabel('Age')
plt.ylabel('Satisfactory Score')
plt.grid(True)
plt.show()
```
